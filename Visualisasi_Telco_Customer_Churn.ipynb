##Processing

```python
#checking missing values
df.isnull().sum()
```

```python
null_value = [col for col in df.columns if df[col].isnull().sum() !=0]
null_value
```

```python
for col in null_value:
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
#Duplicate Data
duplicate = df[df.duplicated(keep = False)]
duplicate
```

```python
df.drop_duplicates(keep='first', inplace=True)
df.reset_index()
df.head()
```
